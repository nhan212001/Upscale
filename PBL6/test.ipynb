```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
```

```python
lr_test = np.load('/PBL/PBL6/Data_train_np/lr_test_images.npy')
hr_test = np.load('/PBL//PBL6/Data_train_np/hr_test_images.npy')
```

```python
g_loss_list = [] # average per epoch list 

for i in range(10):
  # if loss_file.startswith("g_loss"): 
    g_losses = np.load('/PBL/PBL6/save_loss/g_loss_' + str(i+1)+"npy.npy")
    g_loss = np.sum(g_losses) /len(g_losses)
    g_loss_list.append(g_loss)
```

```python
plt.plot(g_loss_list)
```

```python
d_loss_list = [] # average per epoch list 

for i in range(10):
  # if loss_file.startswith("g_loss"): 
    d_losses = np.load('/PBL/PBL6/save_loss/d_loss_' + str(i+1)+"npy.npy")
    d_loss = np.sum(d_losses) /len(d_losses)
    d_loss_list.append(d_loss)
```

```python
plt.plot(d_loss_list)
```

```python
from keras.models import load_model
from numpy.random import randint
generator = load_model('/PBL/PBL6/save_model/gen_e_8.h5', compile=False)
```

```python
# plot all three images
[X1, X2] = [lr_test, hr_test]
# select random example
ix = randint(0, len(X1), 1)
print(len(X1))
# src_image, tar_image = X1[[91]], X2[[91]]
# ix=[476]
src_image, tar_image = X1[ix], X2[ix]
print(ix)

# generate image from source
gen_image = generator.predict(src_image)
print(gen_image.shape, src_image.shape)
plt.figure(figsize=(16, 8))
plt.subplot(231)
plt.title('nh LR')
plt.imshow(src_image[0, :, :, :])
plt.subplot(232)
plt.title('nh thu c')
plt.imshow(gen_image[0, :, :, :])
plt.subplot(233)
plt.title('nh HR')
plt.imshow(tar_image[0, :, :, :])

plt.show()
```

```python
src_image = cv2.imread("/PBL/PBL6/lr_images/5.jpg")
tar_image = cv2.imread("/PBL/PBL6/hr_images/5.jpg")

src_image = np.array(src_image)
src_image = cv2.cvtColor(src_image, cv2.COLOR_BGR2RGB)
src_image = src_image/255.
src_image = cv2.resize(src_image, (32, 32))
src_image = np.expand_dims(src_image, axis=0)

tar_image = np.array(tar_image)
tar_image = cv2.cvtColor(tar_image, cv2.COLOR_BGR2RGB)
tar_image = tar_image/255.
tar_image = np.expand_dims(tar_image, axis=0)

# generate image from source
gen_image = generator.predict(src_image)
print(gen_image.shape, src_image.shape)
plt.figure(figsize=(16, 8))
plt.subplot(231)
plt.title('nh LR')
plt.imshow(src_image[0, :, :, :])
plt.subplot(232)
plt.title('nh thu c')
plt.imshow(gen_image[0, :, :, :])
plt.subplot(233)
plt.title('nh HR')
plt.imshow(tar_image[0, :, :, :])

plt.show()
```
